```python
from sklearn import tree
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
```

```python
wine=load_wine()
```

```python
import pandas as pd
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(wine.data,wine.target,test_size=0.3)
```

```python
clf=tree.DecisionTreeClassifier(criterion='entropy')
```

```python
clf=clf.fit(X_train,Y_train)
score=clf.score(X_test,Y_test)
```

```python
import graphviz
data_dot=tree.export_graphviz(clf,feature_names=wine.feature_names,class_names=['1','2','3'],filled=True,rounded=True)
graph=graphviz.Source(data_dot)
graph
```

```python
import numpy as np
from sklearn.tree import DecisionTreeRegressor
import matplotlib.pyplot as plt

# Create a random dataset
rng = np.random.RandomState(1)
X = np.sort(5 * rng.rand(80, 1), axis=0)
y = np.sin(X).ravel()
y[::5] += 3 * (0.5 - rng.rand(16))

# Fit regression model
regr_1 = DecisionTreeRegressor(max_depth=2)
regr_2 = DecisionTreeRegressor(max_depth=5)
regr_1.fit(X, y)
regr_2.fit(X, y)

# Predict
X_test = np.arange(0.0, 5.0, 0.01)[:, np.newaxis]
y_1 = regr_1.predict(X_test)
y_2 = regr_2.predict(X_test)

# Plot the results
plt.figure()
plt.scatter(X, y, s=20, edgecolor="black",
            c="darkorange", label="sin(x)")
plt.plot(X_test, y_1, color="r",
         label="max_depth=2", linewidth=2)
plt.plot(X_test, y_2, color="b", label="max_depth=5", linewidth=2)
plt.xlabel("x")
plt.ylabel("y")
plt.title("regress tree")
plt.legend()
plt.show()
```
