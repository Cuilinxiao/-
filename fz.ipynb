```python
def readfile(x):
    fr=open(x,encoding='utf-8')
    lines=fr.readlines()
    k=np.zeros((len(lines),3))
    p=[]
    index=0
    fr=open(x,encoding='utf-8')
    fr=fr.readlines()
    for line in fr:
        w=line.strip()#去除每行开头和结尾的特殊符号，默认为空格或制表或换行符
        k[index,:]=w.split('\t')[0:3]
        p.append(w.split('\t')[-1])
        index=index+1
    return k,p
        
        
    
```

```python
y_series=rolling_window(y,10)
```

```python
trainx=y_series[:-valSize-testSize-1]
trainy=y[10:-valSize-testSize]
```

```python
valx=y_series[-valSize-testSize-1:-testSize-1]
valy=y[-valSize-testSize:-testSize]
```

```python
testx=y_series[-testSize-1:-1]
testy=y[-testSize:]
```

```python
trainx=trainx.reshape(trainx.shape[0],trainx.shape[1],1)
valx=valx.reshape(valx.shape[0],valx.shape[1],1)
testx=testx.reshape(testx.shape[0],testx.shape[1],1)
```

```python
import numpy as np
import matplotlib.pyplot as plt 
from keras.models import Model
from keras.layers import Input,Dense,LSTM
from keras.callbacks import ReduceLROnPlateau,ModelCheckpoint,EarlyStopping
from keras.models import load_model
from sklearn.preprocessing import MinMaxScaler
```

```python
inp=Input(shape=(trainx.shape[1],trainx.shape[2]))
x=LSTM(50,return_sequences=True)(inp)
x=LSTM(100,return_sequences=False)(x)
x=Dense(1,activation='linear',name='regression')(x)
model = Model(inp,x)
model.compile(optimizer='rmsprop',
                    loss='mse') 
model.summary() 
```

```python
modelfilepath='best.hdf5'
earlystop=EarlyStopping(monitor='val_loss',patience=20,verbose=1,mode='auto')
checkpoint=ModelCheckpoint(modelfilepath,monitor='val_loss',verbose=1,save_best_only=True,mode='auto')
reduce_lr=ReduceLROnPlateau(monitor='val_loss',patience=10,factor=0.1,min_delta=0.00001,cooldown=0,min_lr=0)
history=model.fit(trainx,trainy,validation_data=(valx,valy),epochs=50,batch_size=32,callbacks=[checkpoint,reduce_lr,earlystop])
```
